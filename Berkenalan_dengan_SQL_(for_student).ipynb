## 2.2 Mendapatkan data unik

```python
# mari kita lihat pemesanan yang diterima oleh perusahaan northwind
query8 = """

SELECT *  FROM Orders

"""
pd.read_sql_query(query8, conn)
```

```python
# kita cek berapa CustomerID unique yang melakukan pemesanan
query9= """

SELECT DISTINCT(CustomerID) FROM Orders

"""
pd.read_sql_query(query9, conn)
```

## 2.3 Menghitung jumlah data di dalam sebuah tabel

```python
# gunakan fungsi COUNT dari untuk menghitung berapa jumlah kategori produk yang di jual
query10 = """

SELECT
COUNT(*) AS jumlah_kategori
FROM Categories

"""
pd.read_sql_query(query10, conn)
```

```python
# kita bisa gabungkan fungsi COUNT dan DISTINCT untuk menghitung jumlah data yang unik di suatu kolom
query11 = """

SELECT
COUNT(DISTINCT(ProductName)) as jumlah_produk
FROM Products

"""
pd.read_sql_query(query11, conn)
```

### 3.3 Pemesanan dan pengiriman

```python
# Berapa banyak jumlah pemesanan yang sudah diterima oleh perusahaan northwind?
# tulis query kamu di sini!
query17 = """

SELECT
COUNT (Quantity) as jumlah_pemesanan
FROM OrderDetails

"""
pd.read_sql_query(query17, conn)
```

```python
# ke negara mana saja pengiriman barang dilakukan oleh northwind ?
# tulis query kamu di sini!
query18 = """

SELECT DISTINCT (ShipCountry) FROM Orders

"""
pd.read_sql_query(query18, conn)
```

```python
# ada berapa nama kapal yang berbeda yang digunakan untuk mengirim produk2 northwind?
# tulis query kamu di sini!
query19 = """

SELECT
DISTINCT(ShipName) as nama_kapal_yang_berbeda
FROM Orders
"""
pd.read_sql_query(query19, conn)
```

```python
# [OPSIONAL] berapa total biaya jasa pengiriman yang dikeluarkan oleh northwind (hint : gunakan kolom freight pada tabel order, serta fungsi aggregasi SUM)
# tulis query kamu di sini!
query8 = """

SELECT SUM(freight)  FROM Orders

"""
pd.read_sql_query(query8, conn)
```
